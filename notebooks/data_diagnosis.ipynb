```python
import pandas as pd

# Load the raw dataset
df = pd.read_csv("../data/jobstreet_all_job_dataset.csv")
print("Dataset loaded successfully")
```

```python
# Check the overall dataset dimensions
print("Total rows:", df.shape[0])
print("Total columns:", df.shape[1])
```

```python
# Display dataset structure and data types
print("\nDataset information:")
df.info()
```

```python
# Preview the first 5 records of the dataset
df.head()
```

```python
# First 10 rows with specific columns only# Display the first 10 records for selected key columns
df[["job_id", "job_title", "company", "location", "type", "salary", "listingDate"]].head(10)
```

```python
# Display 5 random records for selected key columns
df[["job_id", "job_title", "company", "location", "type", "salary", "listingDate"]].sample(5, random_state=42)
```

```python
# Display sample job description text from random records
sample_descriptions = df["descriptions"].sample(3, random_state=42)

for i, desc in sample_descriptions.items():
    print(f"Row index: {i}")
    print(desc)
    print("\n" + "="*100 + "\n")
```

```python
# Calculate the number and percentage of missing values for each column
missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

# Combine missing value results into a summary table
missing_summary = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing Percentage": missing_percent.round(2)
})

# Sort columns by missing value count in descending order
missing_summary = missing_summary.sort_values(by="Missing Count", ascending=False)
missing_summary
```

```python
# Identify the number of unique values
categorical_cols = ["type", "category", "subcategory", "role"]

for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
# Display category values and their frequencies
print("Type value counts:")
print(df["type"].value_counts(dropna=False))

print("\nCategory value counts:")
print(df["category"].value_counts(dropna=False))

print("\nSubcategory value counts:")
print(df["subcategory"].value_counts(dropna=False))

# Display the 20 most frequent values in the role column
print("\nTop 20 role value counts:")
print(df["role"].value_counts(dropna=False).head(20))
```

```python
# Inspect records where employment type is stored in list-like string format
df[df["type"].str.contains(r"\[", na=False)][["job_title", "type"]].head(20)
```

```python
# Inspect records with multiple employment types in the same field
df[df["type"].str.contains(",", na=False)][["job_title", "type"]].head(20)
```

```python
# Check the number of unique location values
print("Number of unique locations:", df["location"].nunique())
```

```python
# Display the 30 most frequent location values
df["location"].value_counts().head(30)
```

```python
# Inspect location values that contain a slash
df[df["location"].str.contains("/", na=False)][["job_title", "location"]].head(20)
```

```python
# Inspect location values that contain a comma
df[df["location"].str.contains(",", na=False)][["job_title", "location"]].head(20)
```

```python
# Check for fully duplicated records across all columns
full_duplicates = df.duplicated().sum()
print("Number of fully duplicated rows:", full_duplicates)
```

```python
# Check for duplicated job_id values
job_id_duplicates = df["job_id"].duplicated().sum()
print("Number of duplicated job_id values:", job_id_duplicates)
```

```python
# Inspect sample non missing salary values
df["salary"].dropna().sample(20, random_state=42)
```

```python
# Display the most frequent salary values
df["salary"].value_counts().head(20)
```

```python
# Check the salary period terms used in the salary text
print("Contains 'month':", df["salary"].str.contains("month", case=False, na=False).sum())
print("Contains 'year':", df["salary"].str.contains("year", case=False, na=False).sum())
print("Contains 'day':", df["salary"].str.contains("day", case=False, na=False).sum())
print("Contains 'hour':", df["salary"].str.contains("hour", case=False, na=False).sum())
```

```python
# Check the symbols used to express salary ranges
print("Contains '-' :", df["salary"].str.contains("-", na=False).sum())
print("Contains '–' :", df["salary"].str.contains("–", na=False).sum())
```

```python
# Inspect unique salary examples to identify currency and format variations
salary_examples = df["salary"].dropna().drop_duplicates()
salary_examples.sample(30, random_state=42)
```

```python
# Inspect records with unusual salary notation or currency format
df[
    df["salary"].notna() &
    df["salary"].str.contains(r"MYR|\$|SGD|\d+\s*[kK]", case=False, na=False, regex=True)
][["job_title", "salary"]].head(30)
```

```python
# Inspect sample values in the listing date column
df["listingDate"].head(10)
```

```python
# Convert listingDate to datetime format for validation and summary checks
listing_date_parsed = pd.to_datetime(df["listingDate"], errors="coerce")

print("Missing after datetime conversion:", listing_date_parsed.isnull().sum())
print("Earliest listing date:", listing_date_parsed.min())
print("Latest listing date:", listing_date_parsed.max())
```

```python
# Review the distribution of records by listing year
listing_date_parsed.dt.year.value_counts().sort_index()
```
